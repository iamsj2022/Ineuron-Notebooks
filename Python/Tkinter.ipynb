# Geometry Management

Tkinter offers the user to change the `geometrc configuration of the widgets` by which we can change the position of the widgets in the parent window. There are `three main types` of geometry management methods. They are:
  * pack() method
  * grid() method
  * place method()

## pack() method

It organizes the widgets in the `order the widgets been packed` before moving it into the parent window. 

### List of parameters:
**expand** - It is used to fill the space of the parent window when set True otherwise not used 
<br>
**fill** - Determines if the widgets packed needs to fill any extra space (X - horizontally, Y - vertically, BOTH - both horizontally and vertically, NONE - default)
<br>
**side** - Determines which side of the parent should a widget pack (TOP - Default, BOTTOM, LEFT, RIGHT)

## grid() method

It organizes the widgets in a `grid or a table like structure` before placing it into the parent widget

### List of parameters:

**column** - The column in which the widget needs to be placed, default 0
<br>
**columnspan** - The number of columns the widget must occupy, default 1
<br>
**padx, pady** - Determines the number of pixels to be padded  outside widgets borders horizontally or vertically
<br>
**ipadx, ipady** - Determines the number of pixels to be padded  inside widgets borders horizontally or vertically
<br>
**row** - The row in which the widget needs to be placed, default 0
<br>
**rowspan** - The number of rows the widget must occupy, default 1
<br>
**sticky** - sticks the widgets in any one of the 8 directions of the compass

### Code1:

Lets now see a grid positioning demonstration

```python
import tkinter as tk
win = tk.Tk()
win.title("Grid Position Demo")
for r in range(3):
   for c in range(4):
      tk.Label(win, text='R%s/C%s'%(r,c)).grid(row=r,column=c)
win.mainloop()
```

## place() method
This method organizes by placing them in a `given specific coordinates`

### List of parameters
**anchor** - Takes any one direction as input out of 8 directions as input NW - North West is default
<br>
**bordermode** − INSIDE (the default) to indicate that other options refer to the parent's inside (ignoring the parent's border); OUTSIDE otherwise.
<br>
**height, width** - Height and width of the widget in pixels
<br>
**relheight, relwidth** − Height and width as a float between 0.0 and 1.0, as a fraction of the height and width of the parent widget. like `0.8 * height of parent widget`
<br>
**relx, rely** − Horizontal and vertical coordinate as a float between 0.0 and 1.0, as a fraction of the height and width of the parent widget.
<br>
**x, y** - Horizontal and Vertical coordinates

### Code
```python
import tkinter as tk

win = tk.Tk()

frame = tk.Frame(win, width=200, height=200)
frame.pack()

label1 = tk.Label(frame, text="(0, 0)", bg="blue")
label1.place(x=0, y=0)

label2 = tk.Label(frame, text="(100, 100)", bg="red")
label2.place(x=100, y=100)

win.mainloop()
```

### Output
Place.png

Here if you see the `blue label` is kept at `(0,0)` coordinate and the `red label` is placed at `(100, 100)`coordinates which is very specific x and y coordinates

## tkMessageBox
It is used to display messageboxes in your applications. It provides a number of functions which you can use to display an appropriate message of your choice

<br>

### Syntax
```python
msgBox = tkMessageBox.FunctionName(title, message [, options])
```
<br>

### Parameters
**FunctionName** − name of the appropriate message box function.

**title** − text to be displayed in the title bar of a message box.

**message** − text to be displayed as a message.

### Function names
* showinfo()
* showwarning()
* showerror ()
* askquestion()
* askokcancel()
* askyesno ()
* askretrycancel ()

### Code

```python
import tkinter as tk
from tkinter import ttk, messagebox

win = tk.Tk()
win.title("Message box")

def assurance():
    answer = messagebox.askyesno("Confirm", "Do you want to close?")
    if answer == True:
        label = tk.Label(win, text="Ok")
        label.pack()
        win.destroy()
    else:
        label = tk.Label(win, text="Canceled")
        label.pack()

button = ttk.Button(win, text="Close", command=assurance)
button.pack()

win.geometry("300x300")
win.mainloop()
```

<br>

### Output

Messagebox.png

Here we have used `askyesorno function` for a `confirmation` whether we need to `close the window or not`. Similarly other functions can be used as per the requirement. Here if we click `yes` then it will `return True` so the `window will close else` it will display a message stating `Canceled`.

import tkinter as tk

border_effects = {
    "flat": tk.FLAT,
    "sunken": tk.SUNKEN,
    "raised": tk.RAISED,
    "groove": tk.GROOVE,
    "ridge": tk.RIDGE,
}

win = tk.Tk()

for relief_name, relief in border_effects.items():
    frame = tk.Frame(win, relief=relief, borderwidth=5)

### Code
```python
import tkinter as tk
      
win = tk.Tk()
win.title("Menu Bar")

menu = tk.Menu(win)
win.config(menu=menu)
filemenu = tk.Menu(menu)
menu.add_cascade(label='File', menu=filemenu)
filemenu.add_command(label='New')
filemenu.add_command(label='Open...')
filemenu.add_separator()
filemenu.add_command(label='Exit', command=win.quit)
helpmenu = tk.Menu(menu)
menu.add_cascade(label='Help', menu=helpmenu)
helpmenu.add_command(label='About')
win.mainloop()
```
